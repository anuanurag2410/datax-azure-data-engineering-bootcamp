### 1️⃣ Why not use storage access keys directly?

Because they provide full access, lack fine-grained control, are hard to rotate, and pose high risk if exposed.

---

### 2️⃣ What is Azure Key Vault?

Centralized service to securely store and manage secrets, keys, and certificates.

---

### 3️⃣ How does ADF connect securely to storage?

Using Managed Identity to retrieve secret from Key Vault or directly access storage via RBAC.

---

### 4️⃣ What is key rotation?

Periodic updating of secrets to minimize exposure and meet compliance requirements.

---

### 5️⃣ How design secure architecture using Key Vault?

Use Managed Identity, store secrets centrally, restrict vault network access, enable soft delete, rotate keys, separate environments.
