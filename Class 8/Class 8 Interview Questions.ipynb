### Q1: What is Azure Data Factory?

✅ Answer:

“Azure Data Factory is a cloud-based data integration service used to orchestrate and automate data movement and transformation workflows across multiple data sources and destinations.”

### Q2: What is Integration Runtime?

✅ Answer:

“Integration Runtime is the execution infrastructure that ADF uses to move data and execute activities, and it enables secure connectivity to cloud/on-prem/private networks.”

### Q3: Linked Service vs Dataset?

✅ Answer:

“Linked Service defines connection to a system, while Dataset defines the specific data reference inside that system like table/file/folder.”

### Q4: Control Flow vs Data Flow?

✅ Answer:

“Control Flow orchestrates workflow logic like branching and loops; Data Flow performs actual data transformations using a Spark-based engine.”

### Q5: How does ADF execute a pipeline?

✅ Answer:

“ADF’s control plane orchestrates the run and the Integration Runtime performs actual execution for data movement and transformations, with monitoring available in ADF Monitor.”
