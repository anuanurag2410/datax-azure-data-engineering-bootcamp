
### Q1: What is Hierarchical Namespace?

Answer:

It enables a true hierarchical file system within Azure Blob Storage, allowing directory-level operations and fine-grained POSIX-style access control.

---

### Q2: Difference between RBAC and ACL?

Answer:

RBAC manages access at the Azure resource level, while ACL controls permissions at file and folder level within ADLS Gen2.

---

### Q3: Why is ACL needed if RBAC exists?

Answer:

RBAC controls entry to storage account, but ACL ensures fine-grained control inside the storage structure.

### Q4: How do you mount ADLS in Databricks?

Answer:

By using a Service Principal authenticated through Azure AD, assigning proper RBAC and ACL permissions, and configuring mount via Spark OAuth credentials.
