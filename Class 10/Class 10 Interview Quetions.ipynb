# Copy Activity – Interview Q&A (ADF)

---

## 1. What is Copy Activity?

Copy Activity is the core component in Azure Data Factory used to move data from a source system to a target system.

In simple terms, whenever we need to ingest data from sources like files, databases, or APIs into storage systems like ADLS or a data warehouse, we use Copy Activity.

It is optimized for data movement and uses Integration Runtime to execute the transfer.

It does not perform heavy transformations; it is mainly focused on efficient data ingestion.

---

## 2. How do you process multiple files?

In real-world scenarios, data usually comes in multiple files rather than a single file.

Instead of hardcoding file names, we use wildcard patterns in Copy Activity.

For example:
*.csv


This allows ADF to automatically pick all CSV files from a folder.

If required, we can also enable recursive to read files from subfolders.

---

## 3. What is Wildcard?

Wildcard is a pattern used to match multiple files dynamically.

Examples:
.csv -> all CSV files
orders_.csv -> files starting with "orders_"


It is useful when file names are not fixed and new files keep arriving.

---

## 4. What is Recursive?

Recursive allows Copy Activity to read files not only from the main folder but also from all nested subfolders.

Example:
/vendor/source/2024/file1.csv
/vendor/source/2025/file2.csv


If recursive is enabled, both files will be picked.

If disabled, only files in the main folder will be processed.

---

## 5. How do you avoid overwrite?

To avoid overwriting data, we do not write to the same location repeatedly.

Instead, we use dynamic folder paths, usually based on date or batch.

Example:
@concat('raw/vendor/', formatDateTime(utcNow(),'yyyy/MM/dd'))


This ensures that every pipeline run writes data to a new location.

This approach is important for traceability and debugging.

---

## 6. How do you handle schema drift?

Schema drift occurs when the structure of incoming data changes, such as new columns being added.

In Copy Activity, we can use auto-mapping to handle schema changes dynamically.

However, in production systems, we usually land raw data first and apply validation and transformation later.

This prevents pipeline failures due to schema changes.

---

## 7. What is Staging?

Staging is a temporary storage layer used during data transfer.

It is especially useful when moving large volumes of data between different systems.

ADF first transfers data into staging and then to the final destination, improving performance and reliability.

---

## 8. How do you optimize performance?

Performance can be optimized in multiple ways:

- Increasing parallel copy to process multiple files simultaneously  
- Using efficient file formats like Parquet instead of CSV  
- Enabling staging for large data transfers  
- Avoiding small file problems by reducing number of small files  

These practices help improve pipeline efficiency and reduce execution time.

# Scenario-Based Questions

---

## Scenario 1

You are receiving 500 files daily. How will you process them?

Answer:

Use wildcard to dynamically pick all files.  
If more control is needed, use ForEach activity.  
Maintain logging to track processed files.

---

## Scenario 2

Pipeline failed after processing some files. What will you do?

Answer:

Maintain metadata to track processed files.  
Restart pipeline from failed state instead of reprocessing all data.

---

## Scenario 3

Same file is coming again. How will you avoid duplicate load?

Answer:

Track file names or use checksum logic.  
Maintain a processed files log and skip already processed files.

---

## Scenario 4

Files are inside multiple nested folders. How will you handle?

Answer:

Enable recursive option to read files from all subfolders.

---

## Scenario 5

You have 1000 small files. What issue can occur?

Answer:

This leads to small file problem, which impacts performance.  
Processing many small files is inefficient.

---

## Scenario 6

How will you design a production-ready ingestion pipeline?

Answer:

- Raw data landing layer  
- Partitioning strategy  
- Logging and monitoring  
- Retry mechanism  
- Fault-tolerant and scalable design  

---

# Final Statement

If I design ingestion using Copy Activity, I will focus not only on moving data but also on making the pipeline scalable, fault-tolerant, and production-ready.
